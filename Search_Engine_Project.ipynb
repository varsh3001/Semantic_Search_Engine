## Reading the data from database

```python
import sqlite3
import pandas as pd
con=sqlite3.connect("eng_subtitles_database.db")
cursor=con.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
print(cursor.fetchall())
```

```python
cursor.execute("PRAGMA table_info('zipfiles')")
columns=cursor.fetchall()
print("The columns of the zipfiles table is:")
for col in columns:
    print(col[1])
```

```python
df=pd.read_sql_query("""SELECT * FROM zipfiles""",con)
df.head()
```

```python
import zipfile
import io
cnt=0
def decode(binary_data):
    global cnt
    cnt+=1
    with io.BytesIO(binary_data) as f:
        with zipfile.ZipFile(f,'r') as zip_file:
            subtitle_content=zip_file.read(zip_file.namelist()[0])
    return subtitle_content.decode('latin-1')
```

```python
df['file_content']=df['content'].apply(decode)
df.head()
```

```python
no_of_rows=len(df.axes[0])
ids=[]
for i in range(1,no_of_rows+1):
    ids.append(i)
df['unique_id']=ids
df
```

## Cleaning the Data

### Taking only 30% of the data

```python
import re
import string

def clean(data):
    lower=data.lower()
    without_html=re.sub(r'<.*?>','',lower)
    pattern = r'\d+\r\n(\d{2}:\d{2}:\d{2},\d{3}) --> \d{2}:\d{2}:\d{2},\d{3}\r\n([\s\S]*?)\r\n\r\n'
    matches = re.findall(pattern, without_html)
    text = [re.sub(r'<\/?i>', '', match[1]) for match in matches]
    #remove punctuation
    cleaned_text = [" ".join(["".join([char for char in word if char not in string.punctuation]) for word in t.split()]) for t in text]
    result = " ".join(cleaned_text)
    return result
```

```python
df1=df.head(24750)
df1
```

```python
copy_df=df1.copy()
```

```python
copy_df['Cleaned Text']=copy_df['file_content'].apply(clean)
```

```python
copy_df
```

```python
def title_clean(title):
    new_title=title.replace("eng.1cd"," ").replace("."," ").strip()
    return new_title
```

```python
copy_df['name']=copy_df["name"].apply(title_clean)
```

```python
copy_df
```

### Chunking the data

```python
import pandas as pd
def chunk_data(row,size=1500):
    data=row['Cleaned Text']
    chunks=[data[i:i+size] for i in range(0,len(data),size)]
    return pd.DataFrame({'name':[row['name']]*len(chunks),'Chunked Data':chunks})
result=pd.concat([chunk_data(i) for _, i in copy_df.iterrows()],ignore_index=True)
print(result)
```

```python
result.to_csv("Chunked_Data.csv")
```
